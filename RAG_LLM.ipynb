```python
from llama_index.llms.huggingface import HuggingFaceLLM
```

```python
document=SimpleDirectoryReader("/content/data").load_data()
document
```

```python
system_prompt="""
Act as a Question and Answer Assistant.
Your goal is to answer questions as accurately as possible based on the instructions and context provided.
"""
#SimpleInputPrompt("<|USER|>{query_str}<ASSISTANT|>")
template=PromptTemplate(system_prompt)
template
```

```python
!!huggingface-cli login
```

```python
import torch

llm = HuggingFaceLLM(
    context_window=2048,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.25, "do_sample": False},
    query_wrapper_prompt=template,
    tokenizer_name="Writer/camel-5b-hf",
    model_name="Writer/camel-5b-hf",
    device_map="auto",
    tokenizer_kwargs={"max_length": 2048},
    # uncomment this if using CUDA to reduce memory usage
    # model_kwargs={"torch_dtype": torch.float16}
)
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
```

```python
!pip install llama-index-embeddings-huggingface
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from llama_index.core import ServiceContext, set_global_service_context

embed_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-mpnet-base-v2"
)
```

```python
!pip install llama-index-embeddings-langchain
```

```python
service_context=ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model
)
```

```python
index=VectorStoreIndex.from_documents(document, service_context=service_context)
```

```python
index
```

```python
query_engine=index.as_query_engine()
```

```python
query_engine.query("What is yolo?")
```
